```python
# prompt: can you import basic library pandas in this workbook

import pandas as pd

# Assuming Salaries.csv is in the current working directory.
# If not, provide the correct path to thdf.iloc[:e file.
try:
    df = pd.read_csv('Salaries.csv')
    print(df)
    print(df.head())
    print(df.tail())
    print(df.columns)
    print(df.shape)
    print(df.dtypes)
    print(df.info())
    pd.set_option('display.max_rows', None)
    print(df)
    print(df.isnull())
    print(df.isnull().sum())
    print(df.describe())
    print(df.describe(include='all'))
    print(df['rank'].value_counts())
    print(df.mean(numeric_only=True))
    print(df['phd'].mean())
    print(df['phd'])
    print(df['phd'].median())
    print(df['salary'].mean())
    print(df['salary'].mode())
    print(df['salary'].median())
    print(df[['phd', 'salary']].mean())
    print(df['salary'].var())
    print(df['salary'].std())
    print(df['salary'].skew())
    print(df['salary'].kurt())
    print(df['salary'].max() - df['salary'].min())
    print(df['salary'].max())
    print(df['salary'].min())
    print(df['salary'].count())
    print(df['rank'].value_counts())
except FileNotFoundError:
    print("Error: Salaries.csv not found. Please ensure the file exists in the current directory or provide the correct path.")
```

```python
# prompt: select first 10 records and first five columns in dataset

import pandas as pd

try:
    df = pd.read_csv('Salaries.csv')
    # Select the first 10 records and the first five columns
    selected_data = df.iloc[0:10, 0:5]
    print(selected_data)
except FileNotFoundError:
    print("Error: 'Salaries.csv' not found. Please make sure the file exists in the current directory or provide the correct path.")
except pd.errors.ParserError:
    print("Error: Could not parse 'Salaries.csv'. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
